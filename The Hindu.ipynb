```python
import os
import requests
from bs4 import BeautifulSoup
from selenium import webdriver
import numpy as np
import pandas as pd
import time
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

driver = webdriver.Chrome(ChromeDriverManager().install())

template = "https://www.thehindu.com/search/?q=article%20370&order=DESC&sort=publishdate&page="
count = 1

driver.get("https://www.thehindu.com/search/?q="+"article+370"+"&order=DESC&sort=publishdate&page=1")
url_set = set()
news_titles = []
news_content = []

while True:
    url = driver.current_url
    count = count+1
    if url in url_set:
        break
    url_set.add(url)
    url = driver.current_url
    html = driver.page_source
    soup = BeautifulSoup(html, 'html5lib')
    articles = soup.find_all("a", class_="story-card75x1-text")  
    print(len(articles))
    
    for n in np.arange(0, len(articles)):
        nurl = articles[n]['href']
        news = requests.get(nurl)
        article_content = news.content
        soup_article = BeautifulSoup(article_content, 'html5lib')
        p = soup_article.find_all('p')
        content = []
        for n in np.arange(0, len(p)):
            content.append(p[n].text)
            final_article = " ".join(content)
        if content:
            print(soup_article.find('title').text)
            news_titles.append(soup_article.find('title').text)
            news_content.append(final_article)
            
    driver.get(template+str(count))
    

```
